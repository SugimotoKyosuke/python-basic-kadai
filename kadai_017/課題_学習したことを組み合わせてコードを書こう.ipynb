```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です。")
    else:
      print(f"{self.name}は大人ではありません。")

taro = Human("太郎", 20)
hanako = Human("花子", 19)

humans = [taro, hanako]

for human in humans:
  human.check_adult()
```
